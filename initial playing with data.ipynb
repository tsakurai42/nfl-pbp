```python
all_pbp_90 = all_pbp.loc[all_pbp['yardline_100'] > 91]
all_pbp_80 = all_pbp.loc[(all_pbp['yardline_100'] > 80) & (all_pbp['yardline_100'] < 91)]
all_pbp_70 = all_pbp.loc[(all_pbp['yardline_100'] > 70) & (all_pbp['yardline_100'] < 81)]
all_pbp_60 = all_pbp.loc[(all_pbp['yardline_100'] > 60) & (all_pbp['yardline_100'] < 71)]
all_pbp_50 = all_pbp.loc[(all_pbp['yardline_100'] > 50) & (all_pbp['yardline_100'] < 61)]
all_pbp_40 = all_pbp.loc[(all_pbp['yardline_100'] > 40) & (all_pbp['yardline_100'] < 51)]
all_pbp_30 = all_pbp.loc[(all_pbp['yardline_100'] > 30) & (all_pbp['yardline_100'] < 41)]
all_pbp_20 = all_pbp.loc[(all_pbp['yardline_100'] > 20) & (all_pbp['yardline_100'] < 31)]
all_pbp_10 = all_pbp.loc[(all_pbp['yardline_100'] > 10) & (all_pbp['yardline_100'] < 21)]
all_pbp_00 = all_pbp.loc[(all_pbp['yardline_100'] > 0) & (all_pbp['yardline_100'] < 11)]
```

```python
each_zones = [all_pbp_90, all_pbp_80, all_pbp_70, all_pbp_60, all_pbp_50, all_pbp_40, all_pbp_30, all_pbp_20, all_pbp_10, all_pbp_00]
```

```python
all_pbp_40.groupby('down').mean()['ep']
```

```python
firstdown = []
seconddown = []
thirddown = []
fourthdown = []
for _ in each_zones:
    test = _.groupby('down').mean()['ep']
    firstdown.append(test[1])
    seconddown.append(test[2])
    thirddown.append(test[3])
    fourthdown.append(test[4])
```

```python
barwidth = 2
r1 = range(90,-1,-10)
r2 = [x + barwidth for x in r1]
r3 = [x + barwidth for x in r2]
r4 = [x + barwidth for x in r3]

plt.bar(r1,firstdown, color="blue", width=barwidth, edgecolor='black', label='1st Down')
plt.bar(r2,seconddown, color="green", width=barwidth, edgecolor='black', label='2nd Down')
plt.bar(r3,thirddown, color="yellow", width=barwidth, edgecolor='black', label='3rd Down')
plt.bar(r4,fourthdown, color="red", width=barwidth, edgecolor='black', label='4th Down')

plt.xlabel('Yards to Endzone')
plt.ylabel('Expected Points')
plt.xticks(range(90,-1,-10))
plt.legend()
plt.show()
```
